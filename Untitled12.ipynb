```python
#lab_03
#Задание 8
#Выполнил: Щукин М.А.
#Группа: ЦИБ-251

def sort_sales_data(sales):
  n = len(sales)
  for i in range(n):
    for j in range(0, n - i - 1):
      if sales[j] < sales[j + 1]:
        sales[j], sales[j + 1] = sales[j + 1], sales[j]
  return sales

if __name__ == "__main__":
    a = [100.5, 50.0, 200.0]
    sort_sales_data(a)
    print(a)
```
